##Funções Merge groupby
agrupe e faça operações pot variável com pd.groupby(<var>)

```python
import pandas as pd

# Dados dos alunos
dados = {'nome': ['Pedro', 'Maria', 'Janaina', 'Marco'],
         'cidade': ['São Paulo', 'Curitiba', 'Brasilia', 'Salvador'],
         'idade': [34, 23, 32, 41],
         'nota': [83.0, 59.0, 86.0, 61.0]}
dfAlunos = pd.DataFrame(dados)

# Novo DataFrame com aprovação
dfAlunos = pd.DataFrame({
    'nome': ['Pedro', 'Maria', 'Janaina', 'Marco'],
    'nota': [78, 52, 89, 64],
    'aprovado(a)': ['Sim', 'Não', 'Sim', 'Sim']
})

# Agrupar por 'aprovado(a)' e calcular a média das colunas numéricas
dfAlunosAprovados = dfAlunos.groupby('aprovado(a)')['nota'].mean()

print('Média de alunos aprovados e reprovados')
print(dfAlunosAprovados)
```

##Indexação de Dados
Indexação(acesso a elementos) com pandas:
.[] -> operador de indexação.
.loc[] -> usada para rótulos.
.iloc[] -> para posição ou bSEda em inteiros

Cria Dataframe com index

```python
area = pd.Series({'Belo Horizonte': 423967, 'São Paulo': 695662,
                  'Curitiba': 141297, 'Rio de Janeiro': 170312,})
pop = pd.Series({'Belo Horizonte': 2423967, 'São Paulo': 4695662,
                  'Curitiba': 1141297, 'Rio de Janeiro': 3170312,})
dados = pd.DataFrame({'area': area, 'população': pop})
dados
```

##Filtro e Manipulações
Filtra os dados com .filter(items=['']) e .filter(regex='')

```python
dados.filter(items=['area'])
```

```python
#procura por expressão que termina com 'e', nas linhas
dados.filter(regex='e$', axis = 0)
```

```python
#procura por expressões que terminam com 'ro', nas linhas
dados.filter(regex='ro$', axis = 0)
```

```python
#procura por expresão que contendo 'ri', naas linhas
dados.filter(like='ri', axis = 0)
```

##Visualização de dados
plota histograma de uma variavel(usuários de internet)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criar o histograma com a distribuição dos dados
plt.figure(figsize=(10, 6))
sns.histplot(demografia['Usuarios de Internet'], kde=True)  # Adiciona a linha de densidade KDE
plt.xlabel('Usuários de Internet')
plt.ylabel('Frequência')
plt.title('Distribuição dos Usuários de Internet')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
visl = sns.distplot(demografia['Usuarios de Internet'], bins = 20)
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criar o displot com a distribuição dos dados
sns.displot(demografia['Usuarios de Internet'], kde=True)  # Adiciona a linha de densidade KDE
plt.xlabel('Usuários de Internet')
plt.ylabel('Frequência')
plt.title('Distribuição dos Usuários de Internet')
plt.show()
```
